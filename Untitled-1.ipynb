```python
import requests

# Votre clé API
api_key = "88382916-d527-45b1-bd27-ccce14a9bac6"

url = "https://api.pokemontcg.io/v2/cards"
headers = {"X-Api-Key": api_key}
page = 1
all_cards = []

for i in range (4):
    params = {
        "page": page,
        "pageSize": 250,
        "select": "name,set,tcgplayer"  # Sélectionne uniquement les champs texte nécessaires
    }
    response = requests.get(url, headers=headers, params=params)

    if response.status_code == 200:
        data = response.json()["data"]
        if not data:
            break
        all_cards.extend(data)
        print(f"Page {page} récupérée avec {len(data)} cartes.")
        page += 1
    else:
        print(f"Erreur: {response.status_code}")
        break

# Affiche le nombre total de cartes extraites
print(f"Nombre total de cartes récupérées : {len(all_cards)}")

# Exemple d'affichage de quelques cartes avec leur prix
for card in all_cards[:10]:
    name = card['name']
    set_name = card['set']['name']
    market_price = card.get('tcgplayer', {}).get('prices', {}).get('normal', {}).get('market', "Non disponible")
    print(f"Nom de la carte: {name}, Set: {set_name}, Prix du marché: {market_price}")
```
